## Campaigns

```python
df['gp:initial_utm_campaign']
```

```python
s = df.groupby('gp:initial_utm_campaign')['user_id'].agg(lambda x:len(x.unique()))
pd.value_counts(s).plot(kind="bar")
```

```python
campaigns = pd.DataFrame({'user_id_count':s})
```

```python
campaigns['percentage_of_traffic_from_campaigns'] = campaigns['user_id_count']*100/campaigns.sum()[0]
```

```python
campaigns['percentage_of_traffic'] = campaigns['user_id_count']*100/df['user_id'].count()
```

### Output of campaigns analysis

```python
campaigns[campaigns['percentage_of_traffic_from_campaigns']>10]
```

```python
campaigns[campaigns['percentage_of_traffic']>6]
```
