## Campaigns Analysis

### Processing

```python
df['gp:initial_utm_campaign']
```

```python
s = df.groupby('gp:initial_utm_campaign')['user_id'].agg(lambda x:len(x.unique()))
pd.value_counts(s).plot(kind="bar")
```

```python
campaigns = pd.DataFrame({'user_id_count':s})
```

```python
campaigns['percentage_of_traffic_from_campaigns'] = campaigns['user_id_count']*100/campaigns.sum()[0]
```

```python
campaigns['percentage_of_traffic'] = campaigns['user_id_count']*100/df['user_id'].count()
```

#### Output of campaigns analysis

```python
campaigns_10pc = campaigns[campaigns['percentage_of_traffic_from_campaigns']>10]
```

```python
campaigns_10pc
```

```python
campaigns_5pc = campaigns[campaigns['percentage_of_traffic']>5]
```

#### Picking data to plot best campaigns

```python
import plotly.express as px
```

```python
px.bar(campaigns_10pc, 
       x=campaigns_10pc.index,         
       y="user_id_count",  
       title=f"Dentre as campanhas - trouxeram mais de 10% dos trials",
       labels={"user_id_count": "#trials"}, # the axis names
       color_discrete_sequence=["blueviolet"], # the colors used
       height=400,
       width=700)
```

```python
px.bar(campaigns_5pc, 
       x=campaigns_5pc.index,         
       y="user_id_count",  
       title=f"Dentre todos os trials - representam mais de 5%",
       labels={"user_id_count": "#trials"}, # the axis names
       color_discrete_sequence=["blueviolet"], # the colors used
       height=400,
       width=700)
```

#### Preparing data to plot best campaigns

```python
list(campaigns_10pc.index)
```

```python
list(campaigns_5pc.index)
```

```python
best_campaigns = list(campaigns_10pc.index) + list(campaigns_5pc.index)
best_campaigns
```

```python
best_campaigns_set = set(best_campaigns)
best_campaigns_set
```

```python
best_campaigns = list(best_campaigns_set)
best_campaigns
```

```python
bc_df = df[df['gp:initial_utm_campaign'].isin(best_campaigns)][['gp:created_at','gp:initial_utm_campaign']]
bc_df
```

```python
bc_df.rename(columns={"gp:created_at": "Data", "gp:initial_utm_campaign": "Campanha"}, inplace=True)
```

```python
bc_df = bc_df.groupby(['Data','Campanha']).size()
```

```python
bc_df = pd.DataFrame(bc_df, columns=['Novos Trials'])
```

```python
bc_df.reset_index(level=1, inplace=True)
```

```python
bc_df
```

```python
bc_df.sort_values(by=['Data'], inplace=True)
```

```python
bc_copy = bc_df.copy()
```

```python
bc_df.columns
```

```python
columns = list(bc_df.columns)
columns.append('Soma Cumulativa dos Trials')
columns
```

```python
bc_df_cumsum = pd.DataFrame(columns=columns)
```

```python
for i in best_campaigns:
    df_i = bc_df[bc_df['Campanha']==i].copy()
    df_i['Soma Cumulativa dos Trials'] = df_i['Novos Trials'].cumsum()
    bc_df_cumsum = pd.concat([bc_df_cumsum,df_i])
    bc_df_cumsum
```

```python
bc_df_cumsum
```
